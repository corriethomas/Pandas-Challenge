```python
#import dependencies
import pandas as pd
import os
```

```python
#read csv, create data frame
purchase_csv = os.path.join("..", "Resources", "purchase_data.csv")
purchase_df = pd.read_csv(purchase_csv)
purchase_df
```

Player Count
- Display the total number of players

```python
#Display total number of players
unique_players = purchase_df["SN"].unique()
total_players = len(unique_players)
total_df = pd.DataFrame({"Total Players": [total_players]
                        })
total_df
```

Purchasing Analysis (Total)
- Run basic calculations to obtain the number of unique items, average price, etc.

```python
#Create variables to store the number of unique items, average price, number of purchases, price total
unique_items = purchase_df["Item ID"].nunique()
unique_items
average_price = purchase_df["Price"].mean()
average_price
price_total = purchase_df["Price"].sum()
price_total
total_purchases = len(purchase_df["Purchase ID"])
total_purchases

#Create data frame to store variables
analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                           "Average Price": [average_price],
                           "Number of Purchases": [total_purchases],
                           "Total Revenue": [price_total]})

#Format columns
analysis_df["Average Price"] = analysis_df["Average Price"].map("${0:,.2f}".format)
analysis_df["Total Revenue"] = analysis_df["Total Revenue"].map("${0:,.2f}".format)
analysis_df
```

Gender Demographics
- Percentage and Count of Male Players
- Percentage and Count of Female Players
- Percentage and Count of Other / Non-Disclosed

```python
#Group original dataframe by gender
grouped_genders = purchase_df.groupby(["Gender"])

#Pull individual players, remove duplicates
genders = grouped_genders["SN"].nunique()
genders

#Create new dataframe to store gender demographics
genders_df = pd.DataFrame({"Total Count": genders,
"Percentage of Players": genders/total_players})

#Format columns
genders_df["Percentage of Players"] = genders_df["Percentage of Players"].map("{:.2%}".format)
genders_df
```

Purchasing Analysis (Gender)
- Run basic calculations to obtain purchase count, avg. purchase price, avg purchase total per person, etc. by gender
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
#Use the grouped genders data frame to create a new data frame/run functions
gender_purchase_analysis_df = pd.DataFrame({"Purchase Count": grouped_genders["Purchase ID"].nunique(),
                                           "Average Purchase Price": grouped_genders["Price"].mean(),
                                           "Total Purchase Value": grouped_genders["Price"].sum(),
                                          "Avg Total Purchase Price per Person": grouped_genders["Price"].sum()/genders_df["Total Count"]})

#Format columns
gender_purchase_analysis_df["Average Purchase Price"] = gender_purchase_analysis_df["Average Purchase Price"].map("${0:,.2f}".format)
gender_purchase_analysis_df["Total Purchase Value"] = gender_purchase_analysis_df["Total Purchase Value"].map("${0:,.2f}".format)
gender_purchase_analysis_df["Avg Total Purchase Price per Person"] = gender_purchase_analysis_df["Avg Total Purchase Price per Person"].map("${0:,.2f}".format)
gender_purchase_analysis_df
```

Age Demographics
- Establish bins for ages
- Categorize the existing players using the age bins
- Calculate the numbers and percentages by age group
- Create a summary data frame to hold the results
- Optional: round the percentage column to two decimal points
- Display age demographics table

```python
#Find the min and max age to determine our bins
min_age = purchase_df["Age"].min()
print(min_age)
max_age = purchase_df["Age"].max()
print(max_age)

#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins
purchase_df["Age Ranges"] = pd.cut(purchase_df["Age"], bins, labels=bin_names, include_lowest=True)
purchase_df

grouped_bins = purchase_df.groupby(["Age Ranges"])

# Calculate the numbers and percentages by age group
bin_df = pd.DataFrame({"Total Count": grouped_bins["SN"].nunique(),
                      "Percentage of Players": (grouped_bins["SN"].nunique())/total_players})

#Format columns
bin_df["Percentage of Players"] = bin_df["Percentage of Players"].map("{:.2%}".format)
bin_df
```

Purchasing Analysis (Age)
- Bin the purchase_data data frame by age
- Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person, etc. in the table below
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
#Used the binned (by age) data frame above to create a new data frame and run calculations
age_purchasing_df = pd.DataFrame({"Purchase Count": grouped_bins["Purchase ID"].nunique(),
                                 "Average Purchase Price": grouped_bins["Price"].mean(),
                                 "Total Purchase Value": grouped_bins["Price"].sum(),
                                 "Avg Purchase Total per Person": (grouped_bins["Price"].sum())/(grouped_bins["SN"].nunique())})

#Format columns
age_purchasing_df["Average Purchase Price"] = age_purchasing_df["Average Purchase Price"].map("${0:,.2f}".format)
age_purchasing_df["Total Purchase Value"] = age_purchasing_df["Total Purchase Value"].map("${0:,.2f}".format)
age_purchasing_df["Avg Purchase Total per Person"] = age_purchasing_df["Avg Purchase Total per Person"].map("${0:,.2f}".format)
age_purchasing_df
```
